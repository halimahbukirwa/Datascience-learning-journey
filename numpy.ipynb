Broadcasting in numpy

```python
scores = [50,60,70,30,25]
```

```python
# To create a numpy array

scores_array = np.array(scores)
scores_array
```

```python
# Change 50,60,70 to 20,40,90 (BROADCASTING)

scores_array[0:3] = [20,40,90]
scores_array
```
