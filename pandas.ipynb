Grouping Data in Pandas

```python
df.groupby(['Uni'])
```

```python
df.groupby(['Uni'])['Age'].mean()
```

```python
df.groupby('Uni')['Age'].mean().sort_values(ascending=False)
```

```python
group = df.groupby('Uni')['Age'].mean().sort_values(ascending = False).reset_index()
group
```

```python
group.columns = ['University','Mean Age']  # Changing column names
group
```
